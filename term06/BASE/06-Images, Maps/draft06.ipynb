```python
#карты в волсе написаны на джаве, но питон тоже позволяет работать с картами
#folium
#фолиум позволяет писовать свои карты
#фолиум повзоляет чето тырить с волса, но для этого нужна старая версия
#pip install folium==0.14
import folium
from folium import plugins
from folium.features import DivIcon
```

```python
map = folium.Map()
map
#карта сразу интерактивная!
# в py будет открываться браузер с картой html
```

```python
#если укажем долготу и ширину, укажем чето конкретное
map = folium.Map(location=[37, 126], zoom_start=7)
map
```

```python
map = folium.Map(location=[37, 126], zoom_start=7, tiles='Cartodb Positron') 
#чето там про attr короче настраивание внешнего вида
map
```

```python
# map.save('name.html') #html!!!
```

```python
folium.Marker(location=[37.56, 126.97]).add_to(map) #маркеры
map #нарисовался Сеул!
```

```python
folium.Marker(location=[37.56, 126.97], popup='The capital of Korea', tooltip='Seoul').add_to(map) #маркеры
#popup - надпись если нажать; tooltip - надпись если навести курсор
map
```

```python
folium.Marker(location=[37.56, 126.97], popup='<b>The capital of Korea</b>', tooltip='Seoul').add_to(map) 
#<b>TEXT</b> -> bold
map
```

```python
# folium.Marker(location=[37.56, 126.97], popup='<b>The capital of Korea</b>', tooltip='Seoul').add_to(map) 
folium.Marker(location=[37.45, 126.70], 
              popup='<b>Korean Airport</b>', 
              tooltip='Incheon',
              icon=folium.Icon(icon='plane', color='red')).add_to(map) #задаем параметры иконки
map
```

```python
#google -- fontawersome(https://fontawesome.com/v4/icons/), getbootstrap.com(https://getbootstrap.com/docs/3.3/components/): про иконки
folium.Marker(location=[37.45, 126.70], 
              popup='<b>Korean Airport</b>', 
              tooltip='Incheon',
              icon=folium.Icon(icon='fire', color='black')).add_to(map) #задаем параметры иконки
map

```

```python
folium.Marker(location=[37.45, 126.70], 
              popup='<b>Korean Airport</b>', 
              tooltip='Incheon',
              icon=folium.Icon(icon='bath', color='red', prefix='fa')).add_to(map) 
            #если берем иконки с сайта fontawersome, надо указывать префикс
map
```

```python
# folium.CustomIcon('...', icon_size=(10, 10)) #можно много че фиксить
#для координат можно сделать следующее:
formatter = 'function(num) {return L.Util.formatNum(num, 5);};' #проще копировать
#функция принимает параметр нум -- ";"" это окончание строки в джаве. форматНум - форматирует там числа
mouseposition = plugins.MousePosition(position='topright', 
                                      separator=' Долгота: ',
                                      num_digits=20,
                                      prefix=' Широта: ', 
                                      lat_formatter=formatter,
                                      lng_first=formatter) #строка, которая будет писать где щас мышка
map.add_child(mouseposition)
map
```

```python
#GeoJSON на волсе (у карты вверху справа) - выделяем весь текст
#заводим файл "1.geojson"
import json
cluster = json.load(open('1.geojson'))
map = folium.Map(location=[50, 90], zoom_start=2)
folium.GeoJson(cluster, zoom_on_click=True, marker=folium.Marker(icon=folium.Icon(color='pink', icon='heart', prefix ='fa'))).add_to(map)
map
```
