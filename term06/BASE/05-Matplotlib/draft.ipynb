```python
fig = plt.figure(figsize=(5, 3))
ax = fig.add_subplot() #координатная ось
ax.set(title='My plot')
# ax.plot(X, Y)
# ax.scatter(X, Y)
ax.bar(X, Y, color='red') #or from google "color picker" (#XXXXX) - HEX
plt.show()
```

```python
import matplotlib.pyplot as plt
def myfunc(power, n):
    X = list(range(1, n + 1))
    Y = [x ** power for x in X]
    plt.plot(X, Y)
    plt.show()
```

```python
myfunc(4, 5)
```

```python
# ipywidgets
# интерактивные графики
from ipywidgets import interact, IntSlider
from IPython.display import display

interact(myfunc, power=IntSlider(min=1, max=5, step=1, value=1), n=list(range(5, 11))) #value: значение power по умолчанию
plt.show()
```
