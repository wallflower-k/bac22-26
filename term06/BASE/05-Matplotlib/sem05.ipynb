#### Задача 1.

Постройте график для функции $x^2 = y$. Поиграйте с настройками внешнего вида графика!

```python
import matplotlib.pyplot as plt 
import numpy as np

x = np.linspace(-10, 10, 100)
y = x ** 2

fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(111)
ax.set(title='My bro', facecolor='#b2bfd4')
ax.grid(True)
ax.plot(x, y, color='r', linestyle='--', linewidth=3)
plt.show()
```

#### Задача 4.

Сгруппируйте значения в табличке с помощью groupby по диапазону цены и сохраните в отдельную переменную. Нарисуйте набор графиков для всех остальных колонок. Можете ли вы по этим графикам сказать, какие характеристики телефонов влияют на цену, а какие вроде бы нет?

#### Задача 5.

Напишите функцию, которая будет строить графики для функции $k * x + b = y$. При этом x - это готовый список чисел, который определен внутри функции (сгенерируйте просто чиселки от 1 до любого числа), а k и b должны передаваться в качестве аргументов функции. 

Добавьте вашей функции интерактивность, позволив пользователю выбирать k и b из списка. 

```python
from ipywidgets import interact, IntSlider
import matplotlib.pyplot as plt

def linear(k, b):
    x = list(range(1, 11))
    y = [k * i + b for i in x]
    fig, ax = plt.subplots(1, 1)
    ax.set_xlim(0, 100)
    ax.plot(x, y)
    plt.grid()
    

ki = int(input('write k: '))
bi = int(input('write b: '))
interact(linear, k=IntSlider(min=1, max=ki, step=1, value=1), b=IntSlider(min=1, max=bi, step=1, value=1)) #value: значение power по умолчанию
plt.show()
```
