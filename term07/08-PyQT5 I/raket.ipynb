```python
import time
def start(n):
    for _ in range(5, 0, -1):
        print(f'Ракета под номером {n} взлетит через секунд: {_}.')
        time.sleep(1)
    print(f'Ракета под номером {n} взлетела.')
    print(time.strftime('%X'))

for i in range(1, 11):
    start(i)

print(time.strftime('%X'))
```

```python
import asyncio
import time
from random import randint

async def factorial(n):
  if n < 0:
    return
  if n == 0:
    return 1
  tasky = asyncio.create_task(factorial(n - 1))
  x = await tasky
  z = n * x
  return z

async def start(n):
    # sec = randint(2, 5)
    taskz = asyncio.create_task(factorial(120))
    x = await taskz
    print(x)
    for _ in range(5, 0, -1):
        print(f'Ракета под номером {n} взлетит через секунд: {_}.')
        await asyncio.sleep(1)
    print(f'Ракета под номером {n} взлетела.')
    print(time.strftime('%X'))

async def main():
    raketas = [asyncio.create_task(start(i)) for i in range(1, 101)]
    for r in raketas:
        await r


print(time.strftime('%X'))

await main()

print(time.strftime('%X'))
```
