Возьмите любой достаточно длинный новостной текст, извлеките из него все именованные сущности и нормализуйте их.

```python
import spacy

nlp = spacy.load('en_core_web_lg')
text = 'In the 1970s, for example, the oil cartel OPEC, led by Saudi Arabia, tried to cut off the supply of oil to the United States. Caused an instant panic, a recession within months. Trump was 27 when those lines of cars snaked around the block. Putin was 21. Xi was 20. Our leaders understand, in their bones, their power comes from fuel.'
doc = nlp(text)

for ent in doc.ents:
    print(f'Entity: {ent.text:30} Label: {ent.label_}')
```

```python
from natasha import (
    Segmenter,
    
    NewsEmbedding,
    NewsMorphTagger,
    NewsSyntaxParser,
    
    Doc
)

segmenter = Segmenter() 
emb = NewsEmbedding()  
morph_tagger = NewsMorphTagger(emb) 
syntax_parser = NewsSyntaxParser(emb)

text1 = 'Государственные нефтеперерабатывающие компании Индии проверяют свои документы, связанные с торговлей нефтью с Россией, чтобы гарантировать отсутствие прямых поставок топлива компаниями "Лукойл" и "Роснефть" на фоне последних санкций Вашингтона против Москвы, передает агентство Рейтер со ссылкой на неназванный источник, знакомый с информацией. В среду Соединенные Штаты анонсировали новый пакет антироссийских санкций, в его рамках были введены меры против "Роснефти", "Лукойла" и их дочерних компаний. Как заявили в Вашингтоне, введение новых санкций обусловлено тем, что у Москвы якобы отсутствует желание вести мирный процесс по Украине. "Государственные нефтеперерабатывающие компании Индии пересматривают свои документы по торговле нефтью с Россией, чтобы гарантировать отсутствие прямых поставок топлива от "Роснефти" и "Лукойла" после того, как обе компании попали под санкции США", — сообщает агентство со ссылкой на источник. Ранее президент США Дональд Трамп говорил, что премьер-министр Индии Нарендра Моди пообещал прекратить покупать нефть у России. Вице-премьер Александр Новак позднее сообщил журналистам, что покупатели российской нефти в Индии продолжат работать с Россией, несмотря на заявления США.'
doc = Doc(text1)
doc.segment(segmenter)
doc.tag_morph(morph_tagger)
doc.parse_syntax(syntax_parser)
sent = doc.sents[0]
# sent.morph.print()
# sent.syntax.print()
```

```python
from natasha import NewsNERTagger

ner_tagger = NewsNERTagger(emb)
doc.tag_ner(ner_tagger)
doc.ner.print()
```

```python
from natasha import MorphVocab

morph_vocab = MorphVocab()

for token in doc.tokens:
  token.lemmatize(morph_vocab)
  print(token.lemma)
```

```python
for span in doc.spans:
    span.normalize(morph_vocab)
   
{_.text: _.normal for _ in doc.spans}
```

```python
import spacy

nlp = spacy.load('ru_core_news_sm')
text2 = 'Государственные нефтеперерабатывающие компании Индии проверяют свои документы, связанные с торговлей нефтью с Россией, чтобы гарантировать отсутствие прямых поставок топлива компаниями "Лукойл" и "Роснефть" на фоне последних санкций Вашингтона против Москвы, передает агентство Рейтер со ссылкой на неназванный источник, знакомый с информацией.'
doc = nlp(text2)
```

```python
import spacy
from spacy import displacy

displacy.serve(doc, style='dep')
```

```python
from nltk import Tree

def to_nltk_tree(node):
    if node.n_lefts + node.n_rights > 0:
        return Tree(node.orth_, [to_nltk_tree(child) for child in node.children])
    else:
        return node.orth_


[to_nltk_tree(sent.root).pretty_print() for sent in doc.sents]
```
