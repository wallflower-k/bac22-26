#### Задача 1. 

Напишите скрипт, который будет запрашивать у пользователя язык и текст и возвращать аудиофайл с этим текстом в озвучке. Пользователь пишет полное название языка!

```python
from gtts import gTTS, lang

lang_dict = lang.tts_langs(
language = input('write language: ').capitalize()
text = input('write ur text: ')
data = dict()

for key, value in lang_dict.items():
    data[value] = key

tts = gTTS(text, lang=data[language])
tts.save('audio_task1.mp3')
```
