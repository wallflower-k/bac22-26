#### Задача 1 (3 балла). 

Ваш друг открыл музыкальный магазин и просит вас помочь ему и написать приложение - базу, где бы можно было фиксировать продажи и закупки. Напишите программу, в которой будет класс магазина: в этом классе должна храниться информация о том, сколько и каких музыкальных инструментов у нас сейчас есть в наличии, а также должны быть два метода - продать инструмент и закупить инструменты. У магазина также есть бюджет. Музыкальный инструмент, наверное, тоже неплохо представить в виде класса: у него должны быть какие-то характеристики, цена, могут быть дополнительные атрибуты (запасной комплект струн). 

Не забудьте написать какой-нибудь код, который будет имитировать рабочий день магазина: покупку и продажу инструментов, продавец (ваш друг) решает, когда ему закупить товар и когда ждать покупателей, которые могут появляться со случайной вероятностью (напоминаю о модуле random). 

```python
class MusicShop:
    budget = 100000
    guitars = []
    pianos = []
    rattles = []

    def __init__(self):
        self.instrumentals = {'pianos': len(self.pianos), 'guitars': len(self.guitars), 'rattles': len(self.rattles)}
    
    def print_inst(self):
        inst = f'Ассортимент:\nВсего пианино: {self.instrumentals["pianos"]}\nВсего гитар: {self.instrumentals["guitars"]}\nВсего трещоток: {self.instrumentals["rattles"]}.'
        print(inst)
    
    def sell(self):
        choice = int(input('Выберите инструмент:\n1 - гитара\n2 - пианино\n3 - трещотка\n'))
        if choice == 1:
            if self.guitars:
                print('Выберите гитару из списка: ')
                for g in range(self.instrumentals['guitars']):
                    print(f'{g + 1}. {self.guitars[g]}')
                user = int(input())
                self.budget += self.guitars[user - 1].price
                del self.guitars[user - 1]
                print('Продано!')
            else:
                print('К сожалению, гитар в наличии нет.')
        elif choice == 2:
            if self.pianos:
                print('Выберите пианино из списка: ')
                for p in range(self.instrumentals['pianos']):
                    print(f'{p + 1}. {self.pianos[p]}')
                user = int(input())
                self.budget += self.pianos[user - 1].price
                del self.pianos[user - 1]
                print('Продано!')
            else:
                print('К сожалению, пианин в наличии нет.')
        elif choice == 3:
            if self.rattles:
                print('Выберите трещотку из списка: ')
                for r in range(self.instrumentals['rattles']):
                    print(f'{r + 1}. {self.rattles[r]}')
                user = int(input())
                self.budget += self.rattles[user - 1].price
                del self.rattles[user - 1]
                print('Продано!')
            else:
                print('К сожалению, трещоток нет.')
        else:
            print('Ошибка: инструмента под таким номером нет.')
        print('Бюджет магазина:', self.budget)
        self.instrumentals = {'pianos': len(self.pianos), 'guitars': len(self.guitars), 'rattles': len(self.rattles)}

    def buy(self):
        if self.budget > 0:
            print('Производим закупку.')
            choice = int(input('Выберите инструмент:\n1 - гитара\n2 - пианино\n3 - трещотка\n'))
            if choice == 1:
                gtype = input('gtype: ')
                spare_strings = int(input('spare strings: 1 or 0'))
                if self.budget - 10000 > 0:
                    if gtype and spare_strings:
                        self.guitars.append(Guitar(gtype=gtype, spare_strings=True))
                    elif gtype:
                        self.guitars.append(Guitar(gtype=gtype))
                    elif spare_strings:
                        self.guitars.append(Guitar(spare_strings=True))
                    else:
                        self.guitars.append(Guitar())
                    self.budget -= 10000
                else:
                    print('Недостаточно средств для закупки.')
            elif choice == 2:
                material = input('material: ')
                ptype = input('ptype: ')
                if self.budget - 15000 > 0:
                    if ptype and material:
                        self.pianos.append(Piano(ptype=ptype, material=material))
                    elif ptype:
                        self.pianos.append(Piano(ptype=ptype))
                    elif material:
                        self.pianos.append(Piano(material=material))
                    else:
                        self.pianos.append(Piano())
                    self.budget -= 15000
                else:
                    print('Недостаточно средств для закупки.')
            elif choice == 3:
                color = input('color: ')
                material = input('material: ')
                if self.budget - 1000 > 0:
                    if color and material:
                        self.rattles.append(Rattle(color=color, material=material))
                    elif color:
                        self.rattles.append(Rattle(color=color))
                    elif material:
                        self.rattles.append(Rattle(material=material))
                    else:
                        self.rattles.append(Rattle())
                    self.budget -= 1000
                else:
                    print('Недостаточно средств для закупки.')
            else:
                print('Ошибка: инструмента под таким номером нет.')
        else:
            print('Недостаточно средств для закупки.')
        print('Бюджет магазина:', self.budget)
        self.instrumentals = {'pianos': len(self.pianos), 'guitars': len(self.guitars), 'rattles': len(self.rattles)}

class Guitar:
    
    def __init__(self, price=10000, brand='Fober', gtype = 'классическая', spare_strings=False):
        self.price = price
        self.brand = brand
        self.gtype = gtype
        self.spare_strings = spare_strings

    def __str__(self):
        dic = {True: 'да', False: 'нет'}
        return f'Гитара {self.brand}, стоимость: {self.price}, тип: {self.gtype}, наличие запасных струн: {dic[self.spare_strings]}'

class Piano:
        
    def __init__(self, price=15000, brand='Zor', material='дерево', ptype='фортепьяно'):
        self.price = price
        self.brand = brand
        self.material = material
        self.ptype = ptype

    def __str__(self):
        return f'Пианино {self.brand}, стоимость: {self.price}, тип: {self.ptype}, материал: {self.material}'

class Rattle:
    
    def __init__(self, price=1000, brand='Tralalelo', color='серый', material='дерево'):
        self.price = price
        self.brand = brand
        self.color = color
        self.material = material

    def __str__(self):
        return f'Трещотка {self.brand}, стоимость: {self.price}, цвет: {self.color}, материал: {self.material}'
```

```python
myshop = MusicShop()
```

```python
print('Симулятор дня.')
while True:
    print('Выберите:\n1 - закупка\n2 - продажа\n3 - наличие инструментов\n0 - закончить рабочий день')
    user = int(input())
    if user == 0:
        print('Рабочий день законился!')
        break
    elif user == 1:
        myshop.buy()
    elif user == 2:
        myshop.sell()
    elif user == 3:
        myshop.print_inst()
    else:
        print('Такой команды нет.')
```

#### Задача 3 (2 балла). 

Ваша подружка попросила вас помочь ей рассортировать ее ужасную гигантскую папку с сохраненками: нужно разложить картинки по подпапкам по нескольким тематикам. По счастью, картинки называются довольно педантичным образом, и по их названиям легко предположить, к какому классу они относятся:

- картинки с мемами про котиков имеют в названии cat или cats (но важно не перепутать с catastrophe и подобными штуками! cat(s) всегда будет отделено дефисом, пробелом или прочерком)
- картинки с мемами про программирование имеют в названии имя паблика - NoP (именно в таком регистре)
- картинки с мемами про диплом содержат thesis. 

Программа должна собрать содержимое папки memes, создать три подпапки для перечисленных выше категорий и рассовать файлы картинок по этим подпапкам. Вам понадобятся библиотеки os, re и, возможно, пригодится shutil (это тоже стандартная библиотека, ее не нужно устанавливать). 

```python
import os, re
#  cat/cats -- NoP -- thesis
# memes -> cats / NoP / thesis
dire = os.listdir('../01 - Exercise/memes')
# print(dire)
patt_cat = re.compile(r'(\b|-|_)cats?(\b|-|_)')
patt_prog = re.compile(r'(\b|-|_)NoP(\b|-|_)')
patt_thes = re.compile(r'(\b|-|_)thesis(\b|-|_)')

cat_files, prog_files, thes_files = list(), list(), list()
for file in dire:
    if re.search(patt_cat, file):
        cat_files.append(file)
    elif re.search(patt_prog, file):
        prog_files.append(file)
    elif re.search(patt_thes, file):
        thes_files.append(file)

for file in cat_files:
    abs_path = os.path.abspath(os.path.join('memes', file))
    os.mkdir('memes/cat')
    new_path = os.path.abspath(os.path.join('memes', 'cat', file))
    os.rename(abs_path, new_path)

for file in prog_files:
    abs_path = os.path.abspath(os.path.join('memes', file))
    os.mkdir('memes/programming')
    new_path = os.path.abspath(os.path.join('memes', 'programming', file))
    os.rename(abs_path, new_path)

for file in thes_files:
    abs_path = os.path.abspath(os.path.join('memes', file))
    os.mkdir('memes/thesis')
    new_path = os.path.abspath(os.path.join('memes', 'thesis', file))
    os.rename(abs_path, new_path)
```
