#### Задача 8. 

Дана строчка, в которой есть двойные, тройные и т.п. пробелы, например такая: 

    '     sd     we       ss dff     s   f'

Удалите из нее все лишние пробелы таким образом, чтобы каждый пробел остался только в одиночном экземпляре. Можно двумя способами!

```python
# решение 1:
s = '     sd      we     ss    dff      s   f'
l = ''
for ind in range(len(s)):
    if s[ind] == ' ' and s[ind + 1] != ' ':
        l += s[ind]
    elif s[ind] != ' ':
        l += s[ind]
print(l)

# решение 2:
s = '     sd      we     ss    dff      s   f'
l = ' '.join(s.split())
if s[0] == ' ':
    l = ' ' + l
print(l)
```

#### Задача 10. 

Оля работает секретарем. Грядет большое собрание, и представители фирм звонят ей один за другим, сообщая, сколько работников от них придут.

В цикле с помощью input() пользователь вводит число работников от каждой фирмы. Как только Оле звонит фирма, отказавшаяся от участия (она посылает 0 своих работников), звонки наконец заканчиваются.

Помогите Оле посчитать, сколько столов понадобится в большом зале, чтобы уместить всех участников: за один стол могут сесть восемь человек.

```python
summ = 0
while True:
    num_emp = int(input())
    if num_emp == 0:
        break
    summ += num_emp
if summ % 8 == 0:
    print(summ // 8)
else:
    print(summ // 8 + 1)
```
